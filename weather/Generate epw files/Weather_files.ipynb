```python
import pandas as pd
import dec
from epw import epw
```

```python
meteo_df=pd.read_csv('./Raw_epw/IOSLOOSL9_from_2016-01-01_to_2016-12-31',index_col='Date_Time',parse_dates=[0,])
meteo_df.head()
```

```python
meteo=meteo_df.resample('H').mean().interpolate()
meteo.head()
```

```python
### To comment if year is NOT 2016
meteo=meteo.drop(pd.date_range(start = '2016-02-29', end = '2016-03-01',freq='H',closed='left'))
```

```python
file=pd.read_csv('./Solar_Radiations/2016.csv')
meteo['Solar_W/m2']=file['GHI'].values
meteo
file['GHI']
file['DNI']
file['DHI']
```

```python
meteo.columns
meteo.index
```

```python
latitude = 59.9138688  # latitude  N
longitude = 10.7522454  # longitude E
elevation = 23  # meters above the sea
```

```python
rad=pd.DataFrame(meteo['Solar_W/m2'].values,index=meteo.index,columns=['GHI'])
```

```python
rad=dec.main(rad,latitude,longitude,elevation)
```
